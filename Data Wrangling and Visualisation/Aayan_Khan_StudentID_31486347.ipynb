### 3. Wrangling Data

### Subsetting The Dataframes

__GDP data__<br>
Only keeping country names and GDP values. Also renaming the 'US dollars)' column to millions USD

```python
df_gdp = df_gdp[['Economy', 'US dollars)']]
df_gdp.rename(columns={"US dollars)":"GDP (millions USD)"}, inplace = True)
```

```python
df_gdp.head()
```

__Population data__

Only keeping country names and population numbers in 2020. Also renaming the column name from 2020 to population

```python
df_population = df_population[['Region, subregion, country or area *', '2020']]
df_population.rename(columns={"2020":"population (thousands)"}, inplace = True)
```

```python
df_population.head()
```

__Vaccination data__

We keep the columns: country, people_fully_vaccinated, total vaccinations (not
from the column total_vaccinations but aggregated from the column
daily_vaccinations as they are inconsistent), and vaccines.
We do this by creating a custom aggregate function and then applying it to the data which is grouped by the difference countries and vaccines. 

After this, the 'country' and 'vaccines' indices are reset using the df.reset_index() funciton which removes the levels from the dataframe we received using the aggregate function so that the dataframes are usable. We also rename the daily_vaccinations columns to total_vaccinations


```python
custom_aggregate = {'people_fully_vaccinated':'max', 'daily_vaccinations':'sum'}
df_vaccination = df_vaccination.groupby(['country', 'vaccines' ]).agg(custom_aggregate)
df_vaccination.reset_index(inplace = True)
df_vaccination.rename(columns = {'daily_vaccinations':'total_vaccinations'}, inplace = True)
```

```python
df_vaccination.head()
```

__3.3: creating list of countries__<br>
We use a list to store the country names as they are mutable (the elements can be changed as we like) and they are also similar to dataframes in structure. It also allows us to use the 'in' command 

```python
countries = ['Indonesia', 'Malaysia', 'Singapore', 'Thailand', 'Philippines','Australia']
```

### 4. Merging dataframes 

Firstly, we use the df.isin() command to filter data for the three dataframes so that we only have data from the countries that we want. Then we just use the pd.merge() command to merge the dataframes by matching the columns containing the country names. 
Finally, the repeated country names are removed

### Filtering data for only the countries we want

```python
df_gdp = df_gdp[df_gdp["Economy"].isin(countries)]
```

```python
df_population = df_population[df_population["Region, subregion, country or area *"].isin(countries)]
```

```python
df_vaccination = df_vaccination[df_vaccination["country"].isin(countries)]
```

### Merging the dataframes

```python
df_combined = df_gdp.merge(df_population, left_on='Economy',right_on='Region, subregion, country or area *')
```

```python
df_combined = df_combined.merge(df_vaccination, left_on='Economy',right_on='country')
```

```python
columns = ['country', 'GDP (millions USD)', 'population (thousands)', 'vaccines', 'people_fully_vaccinated', 'total_vaccinations']
df_combined = df_combined[columns]
```

```python
df_combined
```

### 5. Manage Data Type Issues  

We don't seem to have any data issues as all the data looks valid but we must make sure that all the data types are correct for the dataframe by using the df.dtypes function.
Using this we see that the "GDP (millions USD), population (thousands), vaccines" columns all have object data types and must be changed to integer. Also the people_fully_vaccinated and total_vaccinations can be changed from float to an integer data type as we can only have whole number values for these columns

```python
df_combined.dtypes
```

__GDP Column__<br>
For the GDP (millions USD), we first convert the column to a string using the pd.astype(str) command and then we have to replace the commas with empty characters using the str.replace() command and then we use the pd.to_numeric() command to finally get the column in integer data type. Finally we replace the GDP (millions USD) by the actual population to exactly match the assignment specifications by multiplying the column entries by a million. We also rename this column to remove the (millions)

```python
column = df_combined['GDP (millions USD)']
df_combined['GDP (millions USD)'] = pd.to_numeric(column.astype(str).str.replace(',', ''))
df_combined['GDP (millions USD)'] = df_combined['GDP (millions USD)'] * 10**6
df_combined.rename(columns={'GDP (millions USD)':"GDP"}, inplace = True)
```

__Population Column__<br>
For the population (thousands) column, we first convert the column to a string using the pd.astype(str) command and then we have to replace the sapces with empty characters using the str.replace() command and then we use the pd.to_numeric() command to finally get the column in integer data type. Finally we replace the population (thousands) by the actual population to exactly match the assignment specifications by multiplying the column entries by a thousand. We also rename this column to remove the (millions)

```python
column = df_combined['population (thousands)']
df_combined['population (thousands)'] = pd.to_numeric(column.astype(str).str.replace(' ', ''))
df_combined['population (thousands)'] = df_combined['population (thousands)'] * 10**3
df_combined.rename(columns={'population (thousands)':"population"}, inplace = True)
```

### Question 3

For the final question, you will probably need the non-aggregated data from
“Country-Vaccination.csv”. You are to extract the data that’s related only to
Australia and then plot a line graph on the daily_vaccinations over time. Like
earlier, you are to discard the original total_vaccinations and create the total vaccinations for each day using the cumulative sum of the daily vaccinations (up to that day). Again, plot a line graph to visualise the cumulative vaccinations over time. Explain in what circumstances would the first line graph be useful (if at all) and in what circumstances that the second (cumulative) line graph would be useful?

__Answer:__<br>
I firstly read the file again using pd.read_csv() so that I have raw data. Then I use the equality operator to only have rows where the country is Australia. After that I check the datatypes for each column and notice that the date column would need to be changed to a date datatype so that it can be used for the graph. I do this by using to pd.to_datetime() function. To implement the cumulative vaccinated, I just use a for loop and add the number of vaccinations for the previous day to the current day and then append it to an array which I add to the dataframe at the end.  Then I filter out all the columns I don't need and create a line graph. <br>

Since two doses of the vaccine are needed for the vaccination to be complete, the first line graph of daily_vaccinations would be useful to tell the rate at which people are getting a dose of the vaccine in the country daily. The second line graph would be useful in telling the total number of doses of the vaccines given.

```python
df_vaccination = pd.read_csv('data/country-vaccinations.csv')
df_aus_vaccination = df_vaccination[df_vaccination['country'] == 'Australia']
df_aus_vaccination.reset_index(inplace = True)
df_aus_vaccination.head()
```

```python
df_aus_vaccination.dtypes
```

```python
df_aus_vaccination['date'] = pd.to_datetime(df_aus_vaccination['date'])
df_aus_vaccination.dtypes
```

```python
cumulative_vaccination = [df_aus_vaccination["daily_vaccinations"][1]]
for i in range(1, len(df_aus_vaccination["daily_vaccinations"])):
    cumulative_vaccination.append(df_aus_vaccination["daily_vaccinations"][i] + cumulative_vaccination[i-1])

df_aus_vaccination['total_cumulative_vaccinations'] = cumulative_vaccination

columns = ['date', 'daily_vaccinations', 'total_cumulative_vaccinations']
df_aus_vaccination = df_aus_vaccination[columns]
df_aus_vaccination.tail()
```

```python
df_aus_vaccination.set_index('date').plot(rot=45)
plt.xlabel("Date",size=16)
plt.ylabel("Vaccinations",size=16)
plt.title("Vaccinations over time", size=18)
```
